# ☕ 우리집 카페 AI 사장님 (AI-Assisted Review Management Dashboard)

> **"소상공인의 리뷰 대응 시간 단축과 브랜드 리스크 관리를 위한 인간 참여형(Human-in-the-Loop) AI 솔루션"**
> 본 프로젝트는 부모님이 운영하시는 동네 카페의 실제 비즈니스 문제를 해결하기 위해 개발된, Streamlit 및 LangChain 기반의 리뷰 분석 및 답글 자동 생성 대시보드입니다.

---

## 🎯 1. 프로젝트 기획 배경 (Problem Statement)
* **현황 및 문제점:** 매일 배달 앱과 네이버 플레이스에 쌓이는 고객 리뷰에 답글을 다는 작업은 소상공인에게 큰 시간적 부담입니다. 또한, 감정 노동으로 인해 일관된 톤앤매너를 유지하기 어렵습니다.
* **해결 방안:** AI를 활용해 리뷰의 핵심 요소를 빠르게 분석하고, 사장님이 원하는 말투 스타일별로 답글 초안을 즉시 제안하여 리뷰 대응 프로세스를 효율화합니다.

---

## ✨ 2. 핵심 기능 (Key Features)
1. **미답변 리뷰 선별 수집 (Automation):** Selenium을 활용하여 플랫폼에서 아직 답변이 달리지 않은 '미답변 리뷰'만 필터링하여 대시보드로 가져옵니다. (API 비용 절감 및 유저 피로도 감소)
2. **맞춤형 톤앤매너 대응 (LLM Control):** 사장님의 선택(`친절하게`, `유머러스하게`, `진중하게`)에 따라 GPT-4o-mini가 맞춤형 답글 초안을 생성합니다.
3. **단골 손님 기억 (Session Memory):** `st.session_state`를 활용해 과거 대화 내역을 기억하고, 재방문 고객에게 "지난번에도 찾아주셨는데~"와 같은 맞춤형 멘트를 유도합니다.
4. **인간 참여형 설계 (Human-in-the-Loop):** 완전히 자동 등록되는 방식의 AI 리스크(환각, 잘못된 답변)를 방지하기 위해, AI는 초안만 작성하고 최종 검토 및 등록은 사장님이 직접 하는 안전한 UI를 제공합니다.

---

## 🔒 3. 보안 및 리스크 관리 (Security & Risk Management)
* **API Key 보안:** 민감한 OpenAI API Key는 코드에 하드코딩하지 않고, Streamlit의 `st.secrets` 환경변수로 분리하여 안전하게 관리합니다. `.gitignore` 설정을 통해 GitHub 공유 시 유출을 원천 차단했습니다.
* **플랫폼 제재 및 에러 방어 (Fallback):** 플랫폼의 크롤링 차단이나 웹사이트 구조 변경 리스크에 대응하여, 크롤링 실패 시 사용자가 수동으로 리뷰를 복사·붙여넣기 하여 AI 비서 기능을 지속할 수 있도록 대체 공정(Fallback UI)을 설계했습니다.
